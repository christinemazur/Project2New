```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

```python
csv_file = "Donald-Tweets.csv"
donald_csv_df = pd.read_csv(csv_file)
donald_csv_df
```

```python
for col in donald_csv_df.columns: 
    print(col)
```

```python
clean_csv_df = donald_csv_df[["Date", "Tweet_Text", "Hashtags", "twt_favourites_IS_THIS_LIKE_QUESTION_MARK", "Retweets"]].copy()
clean_csv_df.rename(columns={'Date':'date', 'Tweet_Text': 'tweet_text', 'Hashtags': 'hashtags', 'twt_favourites_IS_THIS_LIKE_QUESTION_MARK': 'twt_favourites_is_this_like_question_mark', 'Retweets':'retweets'}, inplace=True)
clean_csv_df
```

```python
for col in donald_json_file: 
    print(col)
```

```python
clean_json_df = donald_json_file[["created_at", "full_text", "favorite_count", "retweet_count"]].copy()
clean_json_df
```

```python
rds_connection_string = "postgres:Catpuss10!@localhost:5432/trump"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```
